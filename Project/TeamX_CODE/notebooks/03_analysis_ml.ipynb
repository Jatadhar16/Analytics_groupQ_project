# Notebook 3 — Machine Learning Analysis
K-Means Clustering | Random Forest | Linear Regression | ARIMA Forecasting

```python
import sys
sys.path.insert(0, '..')
import pandas as pd
import numpy as np
from src.ml_models import HealthAnalyticsML
ml = HealthAnalyticsML()
```

## 3.1 K-Means Clustering

```python
df = ml.load_profiles()
df_clustered = ml.run_clustering(df)
print('Cluster sizes:')
print(df_clustered['country_cluster'].value_counts().sort_index())
print('\nCluster summary:')
print(ml.results['clustering']['cluster_summary'])
```

## 3.2 Random Forest Regressor

```python
rf_res = ml.run_random_forest(df_clustered)
m = rf_res['metrics']
print(f'R²: {m["r2"]:.4f}')
print(f'RMSE: {m["rmse"]:.4f} years')
print(f'MAE: {m["mae"]:.4f} years')
print(f'CV R²: {m["cv_r2_mean"]:.4f} ± {m["cv_r2_std"]:.4f}')
print('\nTop 5 features:')
print(rf_res['importances'].head(5))
```

## 3.3 Linear Regression

```python
lr_res = ml.run_linear_regression(df_clustered)
print('Coefficient table (significant features p < 0.05):')
coef = lr_res['coef_df']
print(coef[coef['p_value'] < 0.05].to_string())
```

## 3.4 ARIMA Time-Series Forecasting

```python
ts_df = ml.load_time_series()
for country in ['IRL', 'DEU', 'KEN']:
    res = ml.run_arima(ts_df, country_code=country)
    if res:
        print(f'{country}: AIC={res["aic"]:.2f}, ADF p={res["adf_pvalue"]:.4f}')
        print(f'  Forecast 2032: {res["forecast"].values[-1]:.1f} years')
```
